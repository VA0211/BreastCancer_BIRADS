```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('./output/best/prediction_result.csv')
incorrect_predictions = results[results['true_class'] != results['predicted_class']]
incorrect_predictions
```

```python
value = incorrect_predictions['true_class'].value_counts()
plt.figure(figsize=(8, 6))
plt.title('Breast Cancer Finding - Wrongly Predicted Label', fontsize=14)
plt.pie(value, labels=value.index, autopct='%1.1f%%')
plt.show()
```
